```python
# Import Pandas Library

import pandas as pd
```

```python
data=pd.read_csv('ABC_Company_attrition_data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
# Data preprocessing;

#Identify & treat the missing values

data.isnull().sum()
```

```python
# Filling missing values with mean

data['NumCompaniesWorked'] = data['NumCompaniesWorked'].fillna(data['NumCompaniesWorked'].mean())
```

```python
data['EnvironmentSatisfaction'] = data['EnvironmentSatisfaction'].fillna(data['EnvironmentSatisfaction'].mean())
```

```python
data['JobSatisfaction'] = data['JobSatisfaction'].fillna(data['JobSatisfaction'].mean())
```

```python
data['WorkLifeBalance'] = data['WorkLifeBalance'].fillna(data['WorkLifeBalance'].mean())
```

```python
data.isnull().sum()
```

```python
# Check the duplicate records
data[data.duplicated()]
```

```python
# Remove duplicate records. By default keeps the fist value and removes the next duplicate value
data.drop_duplicates()
```

```python
# Correlation

data.corr()
```

```python
#Data Visualisation.Analysing diferent features
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ftr=['Age','TotalWorkingYears','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager','EnvironmentSatisfaction','JobSatisfaction','WorkLifeBalance','JobInvolvement','PerformanceRating']
```

```python
ftr
```

```python
rounded_df = data.round(decimals=1)
fig=plt.subplots(figsize=(20,30))

for p,q in enumerate(ftr):
    plt.subplot(10,1,p+1)
    plt.subplots_adjust(hspace=1.0)
    sns.countplot(x=q,data=rounded_df, hue='Attrition')
    plt.xticks(rotation=90)
```

```python
# Insights:

# Attrition rate is high in the age group of 18 to 26 years
# Employees with 0 to 2 years of total work experience are leaving the company.
# More New joinee attrion is noticed.
# Promotion of the employee plays a significant role in retention
#Years with current Manager-->New joinees & people who are with the same manager for long time (7 to 8 years) are leaving
# Low environment saisfaction, Job satisfaction, & job involvement contributing to attrition


# Recomendations:

# Identify employee high flight risk.
# Create succession slate, train & promote employees.
# Create learning environment, boost new & creative idea generation.
# Hone the leadership skills for managers to be better leaders, for the team where we see high attrition rate.

 
```
